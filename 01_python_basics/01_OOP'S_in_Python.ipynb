```python
#Constructor is a special instance method which is used to initialize properties of our object

#As soon as we create a new object constructor is invoked

#By default a constructor is +nt for every class in python like in C++/JAVA

#In python constructor name is not same as of class like in c++ and java but use a special
#method name for the constructor

#Syntax->
    # def __init__(self,args..):
          #------------BODY------------

class Complex: 
    def __init__(self,real,img):
        self.real=real;
        self.img=img;
    
    def printNo(self):
        if self.img<0:
            print(self.real,'-',-self.img,"i")
        else:
            print(self.real,'+',self.img,"i")
    

c1=Complex(5,-6)
c2=Complex(3,4)

c1.printNo()
c2.printNo()
```

---------------------------------ACCESS MODIFIERS in PYTHON------------------------------

```python
#Public Access Modifier:
    #The members of a class that are declared public are easily accessible from any part of
    #the program. All data members and member functions of a class are public by default.

    #To make an attribute/method public just write it name as normal variable:
        # varName/methodName() -->public in python


        
#Protected Access Modifier:
    #The members of a class that are declared protected are only accessible to a class derived from it.
    
    #Data members of a class are declared protected by adding a single underscore ‘_’
    #symbol before the data member of that class. 
        # _varName/_methodName()-->protected in python

        
        
#Private Access Modifier:
# The members of a class that are declared private are accessible within the class only,
#private access modifier is the most secure access modifier. 

#Data members of a class are declared private by adding a double underscore ‘__’ symbol be
#for the data member of that class. 
    # __varName/__methodName()-->private in python
    

#============================================EXAMPLES====================================
class Programmer:
    
    def __init__(self,name,role,salary):
        self.name = name#public attribute
        self._role = role#protected attribute 
        self.__salary =salary #private attribute 
        
    def changeName(self,newName): #public method
        self.name = newName
    
    def _changeRole(self,newRole): #protected method
        self._role=newRole
    
    def __changeSalary(self,newSalary): #private method
        self.__salary = newSalary
    
    def getSalary(self): 
        print(self.__salary)
        
    def modifySalary(self,newSalary): #public method
        self.__salary = newSalary
    
    def getInfo(self):
        print(self.name)
        print(self._role)
        print(self.__salary) 

person1=Programmer("Deepanshu Bhatt","Ml expert",1000000.7645)
person1.getInfo()

# person1.changeSalary(100000000.29292) #->This will throw error as private

#Public and protected methods can be accessed from outside class in python

print("---------------------------------------------------------------------------------")

person1.changeName("Deepanshu")
person1._changeRole("Deep-Learning Expert")
person1.getInfo()

print("---------------------------------------------------------------------------------")

#similaryly->
print(person1.name) 
print(person1._role)
# print(person1.__salary)->In valid as private attribute


#Now If u want to access/modify private member u can take help of a public method inside class
person1.getSalary()
print("---------------------------------------------------------------------------------")
person1.modifySalary(100000000.72873)
person1.getInfo()
print("---------------------------------------------------------------------------------")

#If u want to access a private attribute/method in python just do->
    # _className__attributeName
print(person1._Programmer__salary) #accessing a private attribute using object 
person1._Programmer__changeSalary(200000000.92231) #accessing private method using object
person1.getInfo()

#But doing above thing make no sense as private attribute/method make no sense if can access form outside
```

```python
# ------------------------------------------------INNER CLASS---------------------------------------------

class Vehicle:
    def __init__(self, veh_type, wheelCount, model, color, maxSpeed):
        self.veh_type = veh_type
        self.wheelCount = wheelCount
        # here i have used var name car to use Car obj
        self.car = self.Car(model, color, maxSpeed)

    def printInfo(self):
        print(self.veh_type, self.wheelCount, end=" ")

    class Car:
        def __init__(self, carName, color, maxSpeed):
            self.carName = carName
            self.color = color
            self.maxSpeed = maxSpeed

        def carInfo(self):
            print(self.carName, self.color, self.maxSpeed)


vec1 = Vehicle("Electric", 4, "Tesla Model 3", "white", 400)
vec1.car.carInfo()  # if i want to call a method of Class Car using vechicle obj

# If i want to create a Car class obj->then also need to use outerClass.InnerClass(arguments for constructor)

car1 = Vehicle.Car("Alto 800", "white", 100)
print(car1.carInfo())
```

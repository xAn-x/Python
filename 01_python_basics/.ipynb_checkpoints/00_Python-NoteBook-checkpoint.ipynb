```python
# Seive of eratothesis->
import math
def seive(n):
    lst=[True for i in range(n+1)]
    end=int(math.sqrt(n))
    
    for i in range(2,end):
        if(lst[i]):
            st=i*i
            for j in range(st,n+1,i):
                lst[j]=False

# Printing primes
    for i in range(2,n+1):
        if(lst[i]):
            print(i,end=' ')
# -------------------------------------------------------------------------
seive(100)
```

```python
#Write a programe to chk if string s and t are anagram

#Method1->
def isAnagram1(s,t):
    len_s,len_t=len(s),len(t)
    
    if(len_s!=len_t):
        return False
    
    dict_s,dict_t={},{}
    for i in range(len_s):
        if(s[i] in dict_s):
            dict_s[s[i]]+=1
        else:
            dict_s[s[i]]=1
        
        if(t[i] in dict_t):
            dict_t[t[i]]+=1
        else:
            dict_t[t[i]]=1
    
    return dict_s==dict_t

#Method2->
def isAnagram2(s,t):
    len_s,len_t=len(s),len(t)
    
    if(len_s!=len_t):
        return False
    
    dict_s,dict_t={},{}
    for i in range(len_s):
        #This get function chk if s[i] is in dict_s if it is it take it's value else it take default val
        dict_s[s[i]]=1+dict_s.get(s[i],0)
        dict_t[t[i]]=1+dict_t.get(t[i],0)
        
    
    return dict_s==dict_t

#Method3->
from collections import Counter 
def isAnagram3(s,t):
    # The Counter holds the data in an unordered collection, just like hashtable objects.
    # The elements here represent the keys and the count as values
    
    #To use this import Counter from Collections
    return Counter(s)==Counter(t)

s,t=input().split(" ")

print(isAnagram3(s,t))
```

```python
# A Set is an unordered collection data type that is iterable, mutable and has no duplicate 
# Python’s set class represents the mathematical notion of a set.
# The major advantage of using a set, as opposed to a list, is that it has a highly optimized
# method for checking whether a specific element is contained in the set.
# This is based on a data structure known as a hash table.
# Since sets are unordered, we cannot access items using indexes like we do in lists. 


# Syntax->set={keys.....}
s={1,"Deepanshu","Hell No",2.334}
print(s)

# s[0] ->make no sense as set are unordered collections 

# To chk if a ele in set use in
print("Deepanshu" in s)

#For each loop
for ele in s:
    print(ele,end=' ')

#len()-return length of set
print('\n',len(s))


# From any data-structure to set->
lst=range(10)
my_set=set(lst)
print(my_set)
```

```python
# In python set are kind a similar to set in maths and support a no of mathematical functionalities

a={1,2,3,4}
b={3,4,5,6,7}

# a.intersection(b)-return a intersection b
print(a.intersection(b)) # same as a & b
print("intersection: ",a & b)


# a.union(b)-return a union b
print(a.union(b)) # same as a | b
print("union: ",a | b)



# a.difference(b)-return a - b =>elements in a but not in b
print(a.difference(b))


# a.symmetric_difference(b)-return  symmetric difference of a and b ->i.e return a∪b - a∩b
print(a.symmetric_difference(b))


#U can update set to intersection/symmetric difference/difference etc as shown->
a.symmetric_difference_update(b)#->update a to symmetric difference of a and b

#U can't do union update
print(a)


#A.issubset(B)->chk if A is subset of B
print(a.issubset(b))

# Like issubset() we have->
                    #->issuperset()
                    #->isdisjoint()
```

```python
# enumerate(iterable, start=0) -->when dealing with iterators, we also get a need to keep a count of iterations. Python eases the programmers’ task by providing a built-in function enumerate() for this task. 
# enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly for loops or converted into a list of tuples using the list() method.

for i in enumerate("Deepanshu"):
    print(i)
    
obj=enumerate(["Deepanshu","Golem","sAsukeee"])

print(type(obj))

print(list(obj))


# We can use both counter and value while iterating over the enumerate
print("--------------------------------------------------------------------")
# first var is for counter 2nd for value
for i,s in enumerate(("Hello","hola","Namastey")):
    print(i,s)
```

```python
# Python zip() method takes iterable or containers and returns a single iterator object, having mapped values from all the containers. 

# It is used to map the similar index of multiple containers so that they can be used just using a single entity. 
```

```python
# Python zip() example
# Example 1: Python zip two lists

name = [ "Manjeet", "Nikhil", "Shambhavi", "Astha" ]
roll_no = [ 4, 1, 3, 2 ]
 
# using zip() to map values
mapped = zip(name, roll_no)
 
print(set(mapped))
```

```python
# Example 2: Python zip enumerate

names = ['Mukesh', 'Roni', 'Chari']
ages = [24, 50, 18]
 
for i, (name, age) in enumerate(zip(names, ages)):
    print(i, name, age)
```

```python
# Example 3: Python zip() Dictionary

stocks = ['reliance', 'infosys', 'tcs']
prices = [2175, 1127, 2750]
 
new_dict = {stocks: prices for stocks,
            prices in zip(stocks, prices)}
print(new_dict)
```

```python
# Unzipping means converting the zipped values back to the individual self as they were. This is done with the help of “*” operator.
 
# initializing lists
name = ["Manjeet", "Nikhil", "Shambhavi", "Astha"]
roll_no = [4, 1, 3, 2]
marks = [40, 50, 60, 70]
 
# using zip() to map values
mapped = zip(name, roll_no, marks)
 
# converting values to print as list
mapped = list(mapped)
 
# printing resultant values
print("The zipped result is : ", end="")
print(mapped)
 
print("\n")
 
# unzipping values
namz, roll_noz, marksz = zip(*mapped)
 
print("The unzipped result: \n", end="")
 
# printing initial lists
print("The name list is : ", end="")
print(namz)
 
print("The roll_no list is : ", end="")
print(roll_noz)
 
print("The marks list is : ", end="")
print(marksz)
```
